```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Path to the audio file
audio_file_path = "/content/2023 The Year in Review.mp3"

# Load the audio file
y, sr = librosa.load(audio_file_path, sr=None)  # sr=None preserves original sampling rate

# Generate the mel spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)  # Convert to decibel scale for better visualization

# Plot the mel spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(S_dB, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram for Podcasts')
plt.xlabel('Time (s)')
plt.ylabel('Mel Frequency (Hz)')
plt.tight_layout()
plt.show()
```
