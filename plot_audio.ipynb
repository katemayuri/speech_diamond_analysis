```python
#running code
#code to plot distribution of audio files wrt audio duration and audio files count for Rapaport podcasts
import os
from pydub import AudioSegment
import matplotlib.pyplot as plt
import numpy as np

# Path to the folder containing audio files
audio_folder = "/mnt/c/Users/mayur/Desktop/Dal - Fall 2024/CSCI 6518 - Deep Speech/Project/Data Set Analysis/Data-Set/Audio/Rapaport Podcasts/Audio"
audio_lengths = []

# Function to get audio duration
def get_audio_duration(filepath):
    audio = AudioSegment.from_file(filepath)
    return len(audio) / 1000  # Duration in seconds

# Process all audio files
for file in os.listdir(audio_folder):
    if file.endswith(".mp3") or file.endswith(".mp4"):
        filepath = os.path.join(audio_folder, file)
        duration = get_audio_duration(filepath)
        audio_lengths.append(duration)

# Plot histogram of audio lengths
plt.figure(figsize=(10, 6))
plt.hist(audio_lengths, bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Audio Lengths for Rapaport Podcasts")
plt.xlabel("Duration (seconds)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Print descriptive statistics
print("Total Audio Files:", len(audio_lengths))
print("Total Duration (minutes):", np.sum(audio_lengths) / 60)
print("Average Duration (seconds):", np.mean(audio_lengths))
print("Median Duration (seconds):", np.median(audio_lengths))
print("Standard Deviation:", np.std(audio_lengths))
```
