Code to plot Amplitude histogram of single Audio file

```python
#code to plot Amplitude histogram of single Audio file
import librosa
import numpy as np
import matplotlib.pyplot as plt

def plot_amplitude_histogram(file_path):
    try:
        # Load the audio file
        y, sr = librosa.load(file_path, sr=None)
        
        # Plot the amplitude histogram
        plt.figure(figsize=(10, 6))
        plt.hist(y, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
        plt.title(f'Amplitude Histogram for Rapaport Podcasts', fontsize=16)
        plt.xlabel('Amplitude', fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.grid(True, alpha=0.5)
        plt.show()
        
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# Example usage
file_path = "/Audio/Rapaport Podcasts/Audio/A Crisis in Green Diamonds.mp3" #one audio file
plot_amplitude_histogram(file_path)
```

Code to plot Spectral histogram of single Audio file

```python
#code to plot Spectral histogram of single Audio file
import librosa
import numpy as np
import matplotlib.pyplot as plt

def plot_spectral_histogram(file_path):
    try:
        # Load the audio file
        y, sr = librosa.load(file_path, sr=None)
        
        # Compute the Short-Time Fourier Transform (STFT) to get the frequency components
        D = np.abs(librosa.stft(y))  # Get the magnitude of the STFT
        
        # Convert to dB scale
        S_db = librosa.amplitude_to_db(D, ref=np.max)
        
        # Plot the histogram of the frequency components
        plt.figure(figsize=(10, 6))
        plt.hist(S_db.flatten(), bins=50, color='skyblue', edgecolor='black', alpha=0.7)
        plt.title(f'Spectral Histogram for Rapaport Podcasts', fontsize=16)
        plt.xlabel('Frequency (dB)', fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.grid(True, alpha=0.5)
        plt.show()
        
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# Example usage
file_path = "/Audio/Rapaport Podcasts/Audio/A Crisis in Green Diamonds.mp3"  # One AUdio file
plot_spectral_histogram(file_path)
```
